```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("Automobile.csv")

print(df.head())

print(df.info())
```

```python
# 1) mpg vs weigth

plt.figure()
plt.scatter(df['weight'], df['mpg'])
plt.xlabel('Peso')
plt.ylabel('Millas por galón')
plt.title('Relación entre peso y consumo')
plt.show()
```

```python
# 2) Histograma de consumo

plt.figure()
plt.hist(df['mpg'], bins=20)
plt.xlabel('mpg')
plt.ylabel('Frecuencia')
plt.title('Distribución del consumo de combustible')
plt.show()
```

```python
# 3) Boxplot
plt.figure()
plt.boxplot([df['cylinders'], df['origin']], labels=['Cylinders', 'Origin'])
plt.title('Boxplot de Cylinders y Origin')
plt.show()
```

```python
#En este punto busque como solucionarlo pero no pude
plt.figure()
tabla = pd.crosstab(df['model_year'], df['origin'])
tabla.plot(kind='bar')
plt.xlabel('Año del modelo')
plt.ylabel('Cantidad de autos')
plt.title('Distribución de autos por año y origen')
plt.legend(title='Origen')
plt.show()
```
