```python
import numpy as np

# 1
ceros = np.zeros(10)
print(ceros)

# 2
nums = np.arange(1, 21)
print(nums)

# 3
pares = np.arange(2, 51, 2)
print(pares)

# 5
matriz_random = np.random.rand(4, 4)
print(matriz_random)

# 6
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
print(a + b)

# 7
m1 = np.random.randint(1, 11, (3, 3))
m2 = np.random.randint(1, 11, (3, 3))
resultado = m1 @ m2
print(resultado)

# 8
valores = np.random.randint(1, 101, 15)
print(valores)
print("Max:", valores.max())
print("Min:", valores.min())
print("Prom:", valores.mean())

# 9
matriz = np.arange(1, 26).reshape(5, 5)
print(matriz)
print("Diagonal:", np.diag(matriz))

# 10
mat = np.random.randint(1, 21, (5, 5))
print(mat)
print("Suma filas:", mat.sum(axis=1))
print("Suma columnas:", mat.sum(axis=0))
```
